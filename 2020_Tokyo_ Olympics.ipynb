# Importing Relevant Liabraries

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
import seaborn as sns
sns.set()
```

# Loading Data

```python
Athletes=pd.read_excel('Athletes.xlsx')
Coaches=pd.read_excel('Coaches.xlsx')
EntriesGender=pd.read_excel('EntriesGender.xlsx')
Medals=pd.read_excel('Medals.xlsx')
Teams=pd.read_excel('Teams.xlsx')
```

# Discriptive Graphs

# Medals won By Country
















```python
medals_country = Medals.loc[:,'Team/NOC']
medals_Gold = Medals.loc[:,'Gold']
medals_Silver = Medals.loc[:,'Silver']
medals_Bronze = Medals.loc[:,'Bronze']
fig=plt.figure()
ax=fig.add_axes([0,0,1.5,1])
fig.set_size_inches(18, 10)
ax.bar(medals_country,medals_Bronze,color = '#CD7F32')
ax.bar(medals_country,medals_Silver,bottom=medals_Bronze,color = '#C0C0C0')
ax.bar(medals_country,medals_Gold,bottom=medals_Silver+medals_Bronze,color = '#FFD700')
plt.xticks(rotation='vertical',size=20)
plt.title('Medals Won By Each Country',size=20)
plt.legend(['Gold','Silver','Bronze'])
plt.show()
```

### USA scored most of the medals with over 100 medals.















# Team Participation VS Discipline












```python
Discipline = Teams.loc[:,'Discipline']
sns.countplot(Discipline)
plt.xticks(rotation='vertical')
plt.title('Team Participations VS Discipline')
plt.show()
```

### Swimming seems to have the most number of teams.









# Athletes VS Discipline

```python
Athletes_dis = Athletes.loc[:,'Discipline']
sns.countplot(Athletes_dis)
plt.title('Athletes VS Discipline')
plt.xticks(rotation='vertical')
plt.rcParams["figure.figsize"] = (200,20)
plt.show()
```

### This bar graph shows that the discipline athletics had the most number of athletes.

# Coaches VS Discipline

```python
Coaches_dis = Coaches.loc[:,'Discipline']
sns.countplot(Coaches_dis)
plt.title("Coaches VS Discipline")
plt.xticks(rotation='vertical')
plt.rcParams["figure.figsize"] = (10,3)
plt.show()
```

### Here we infer that athletes from Basketball and Artistic Swimming had most number of coaches
